```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
lst1 = []
url = "https://www.analyticsvidhya.com/blog/2021/05/20-questions-to-test-your-skills-on-k-nearest-neighbour/"
page_source = requests.get(url)
soup = BeautifulSoup(page_source.content, "html.parser")
answers = soup.find_all(["h2", "p"])
for answer in answers:
    lst1.append(answer.text)
lst1
```

```python
lst_new1 = lst1[7:105]
lst_new1
```

```python
pattern=r"^\d+\.[\w\d\s]+\?"
flag=2
str1=""
ques=[]
ans=[]
j=0
for i in lst_new1:
    j=j+1
    w=re.findall(pattern,i)
    #print(w)
    if len(w)>0:
        ques.append(str(i))
        if len(ques)==flag:
            ans.append(str1)
            str1=""
            flag=flag+1
    else:
        str1=str1+str(i)
ans.append(str1)
print(len(ques))
print(len(ans))
for i in range(len(ans)):
    ans[i]=ans[i].replace('\n'," ")
for i in range(len(ans)):
    #ques[i]=ques[i].replace('\n'," ")
    ques[i]=re.sub(r"[0-9 ]+\."," ",ques[i])
    
df=pd.DataFrame({"Questions":ques,"Answer":ans})
pd.set_option("display.max_colwidth", None)
df.head()
```
