```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
lst1= []
url = "https://indianexpress.com/todays-paper/"
page_source = requests.get(url)
soup = BeautifulSoup(page_source.content, "html.parser")
answers = soup.find_all(['strong'])
for answer in answers:
    lst1.append(answer.text)
lst1
```

```python
lst_new = lst1[6:66]
lst_new 
```

```python
import pandas as pd 
```

```python
The_Indian_Express = pd.DataFrame(lst_new, columns=['News Headings'])
```

```python
The_Indian_Express.head()
```

```python
lst2 = []
failed_page = []
for p in range(1,15):
    link = "https://www.ndtv.com/india/page-"+str(p)
    page = requests.get(link)
    if(str(page) == "<Response [200]>"):
        soup = BeautifulSoup(page.content,'html.parser')
        titles = soup.find_all(['h2'])
        for i in titles:
            lst2.append(i.text)
    else:
        failed_page.append(p)
print(lst2)
print(failed_page)
```

```python
NDTV = pd.DataFrame(lst2, columns=['News Headings'])
```

```python
NDTV
```

```python
lst3 = []
failed_page = []
for p in range(2,16):
    link = "https://www.thehindu.com/news/national/?page="+str(p)
    page = requests.get(link)
    if(str(page) == "<Response [200]>"):
        soup = BeautifulSoup(page.content,'html.parser')
        titles = soup.find_all([{'class':'Other-StoryCard-heading'},'h3'])
        for i in titles:
            lst3.append(i.text)
    else:
        failed_page.append(p)
print(lst3)
print(failed_page)
```

```python
lst00 = lst3[26:]
lst00
```

```python
for i in range(len(lst00)):
    lst00[i]=lst00[i].replace('\n',"").strip()
```

```python
lst00
```

```python
THE_HINDU = pd.DataFrame(lst00, columns=['News Headings'])
```

```python
THE_HINDU
```

```python
News = pd.concat([The_Indian_Express, NDTV, THE_HINDU]).reset_index(drop=True)
pd.set_option("display.max_colwidth", None)
```
