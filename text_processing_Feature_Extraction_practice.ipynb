```python
import pandas as pd
book = pd.read_csv("apple.txt", error_bad_lines=False)
```

```python
book
```

```python
book = [x.strip() for x in book.x]  # remove both the leading and trailing charactors
```

```python
book
```

```python
book = [x for x in book if x] # remove empty string because they are considered in python as false
book[0:10]
```

```python
#joining the list into one string/test
text = ' '.join(book)
text
```

```python
#Punctuation
import string # special operations on string
no_punc_text = text.translate(str.maketrans('', '', string.punctuation))
# with arg (x, y, z) where x $ y must be equal-length strings. char x replace by char y. and z is string
no_punc_text
```

```python
#Natural Language Tool kit (NLTK) its library for NLP
import nltk
#Tokenization
from nltk.tokenize import word_tokenize #sent_tokenize
text_token = word_tokenize(no_punc_text)
print(text_token[0:50])
```

```python
len(text_token)
```

```python
#stopwards
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Remove stopwords
my_stop_words = stopwords.words('english')
my_stop_words.append('the')

no_stop_tokens = [w for w in text_token if not w in my_stop_words]
print(no_stop_tokens[0:40])
```

```python
len(my_stop_words)
```

```python
#my_stop_words
```

```python
len(no_stop_tokens)
```

```python
# Normalize the data
lower_words = [x.lower() for x in no_stop_tokens]
print(lower_words[0:25])
```

```python
#Stemming
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
stemmed_tokens = [ps.stem(word) for word in lower_words]
print(stemmed_tokens[0:40])
```

```python
#Nlp english language model of spacy library
import spacy # language models
nlp = spacy.load("en_core_web_sm")
```

```python
# lemma being one of them but mostly POS which will follow later
#doc = nlp(' '.join(no_stno_stop_tokens))
doc = nlp(' '.join(lower_words))
print(doc[0:40])
```

```python
lemmas = [token.lemma_ for token in doc]
print(lemmas[0:40])
```

# Feature Extraction

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(lemmas)
help(CountVectorizer)
```

```python
print(vectorizer.vocabulary_)
```

```python
print(vectorizer.get_feature_names()[50:100])
print(X.toarray()[50:100])
```

```python
print(X.toarray().shape)  # here we can see no. of features(too larege)
```

## Let's see how can bigram and trigram can be include here

```python
vectorizer_ngram_range = CountVectorizer(analyzer='word', ngram_range=(1,3), max_features=100)
BoW_matrix_ngram = vectorizer_ngram_range.fit_transform(book)
```

```python
print(vectorizer_ngram_range.get_feature_names())
print(BoW_matrix_ngram.toarray())
```

# TFidf vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer_n_gram_max_feature = TfidfVectorizer(ngram_range=(1, 3), analyzer='word', norm='l2', max_features=500)
TFidf = vectorizer_n_gram_max_feature.fit_transform(book)
print(vectorizer_n_gram_max_feature.get_feature_names())
print(TFidf.toarray())
```

```python
# Image packege
from wordcloud import WordCloud, STOPWORDS
# define  a function to plot wordcloud
def plot_cloud(wordcloud):
    plt.figure(figsize=(40,30)) # set figure size
    plt.imshow(wordcloud)       #Display image
    plt.axis('off')             #NO axis detials
```

```python
# Denerate wordcloud
stopwords = STOPWORDS
stopwords.add('laptop')
stopwords.add('product')
wordcloud=WordCloud(width=3000,height=1000,background_color='black',max_words=100,colormap='Set2',stopwords=stopwords).generate(text)
#plot
plot_cloud(wordcloud)
```
